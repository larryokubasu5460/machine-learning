
Predict canada's per capita income in year 2020. There is an exercise folder here on github at same level as this notebook, download that and you will find canada_per_capita_income.csv file. Using this build a regression model and predict the per capita income fo canadian citizens in year 2020</p>

```python
import  pandas as pd
import numpy as np
%matplotlib inline
import sklearn.metrics as sm
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
df=pd.read_csv('canada_pci.csv')
df.head()
```

```python
df=df.rename(columns={'per capita income (US$)':'income'})
df.head()
```

```python
plt.subplots(figsize=(12,9))
plt.xlabel('Year')
plt.ylabel('Per capita Income (US$)')
plt.scatter(df.year,df.income, color='red')
```

```python
plt.subplots(figsize=(12,8))
plt.xlabel('Year')
plt.ylabel('Income')
plt.title("Canada")
plt.xticks(df.year, rotation='vertical')
plt.plot(df.year, df.income)
```

```python

X=df.drop('income',axis='columns')
X.head()
```

```python
y=df.income
y.head()
```

```python
model=LinearRegression()
model.fit(X,y)
```

```python
model.predict([[2020]])
```

```python
model.coef_
```

```python
model.intercept_
```
